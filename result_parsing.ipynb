```python
import requests
import json
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
```

```python
url = 'https://charusat.edu.in:912/UniExamResult/frmUniversityResult.aspx'
headers = {'content-type': 'application/x-www-form-urlencoded', 'Accept': '*/*', 'connection': 'keep-alive'}
```

```python
with open('data.json') as f:
    req_data = json.load(f)
req_data['txtEnrNo'] = ""

# Removing the id cause that would be dynamic
```

```python
payload_initial = ''

for key in req_data:
    value = req_data[key]
    payload_initial += key + "=" + value + "&"
payload_initial = payload_initial[:-1]
payload_initial

# POST request payload shit's iniial part. only id needs to be appended
```

```python
df = pd.DataFrame()
# defining the DataFrame in which the data is going to get stored
```

```python
gone_students_numbers = {34, 93, 104, 141, 144, 148, 170}
d2d_students_starts_from = 160
total_number_of_students = 185
```

```python
def get_ID(n):
    if n in gone_students_numbers:
        return ""
    if n < d2d_students_starts_from:
        return f"19DCS{n:03d}"
    elif n >= d2d_students_starts_from:
        return f"D20DCS{n:03d}"
    return ""

def get_all_IDS():
    return list(filter(None, map(get_ID, range(1, total_number_of_students))))
```

```python
get_all_IDS() # trial
```

```python
def get_payload_text(id_num):
    return payload_initial+get_ID(id_num)
def get_html_response(id_num):
    return requests.post(url, data=get_payload_text(id_num), headers=headers).text
def get_html_response_soup(id_num):
    return BeautifulSoup(get_html_response(id_num), "html.parser")
```

```python
soup = get_html_response_soup(89)
# trial and for then feeding into trials cause I donn wanna be requesting for every function trial
```

```python
student_info = {}
```

```python
def get_basic_student_info(soup):
    student_info['ID'] = soup.find("span", {"id": "uclGrd1_lblExamNo"}).text
    student_info['name'] = soup.find("span", {"id": "uclGrd1_lblStudentName"}).text
    # student_info['programme'] = soup.find("span", {"id": "uclGrd1_lblProgramme"}).text
    # student_info['month'] = soup.find("span", {"id": "uclGrd1_lblMnthYr"}).text.split(' ')[0]
    # student_info['year'] = soup.find("span", {"id": "uclGrd1_lblMnthYr"}).text.split(' ')[1]
    # student_info['sem'] = soup.find("span", {"id": "uclGrd1_lblSemester"}).text
```

```python
get_basic_student_info(soup) # Trial
student_info
```

```python
sub_codes = set()
sub_names = set()
subs = []
subs_abbr = []
```

```python
def get_subs_info(soup):
    global sub_codes, sub_names, subs, subs_abbr
    sub_codes = set()
    sub_names = set()
    subs = []
    subs_abbr = []

    tds = soup.find_all("table", {"id": "uclGrd1_grdResult"})[0].find_all("td", {"class": "Newtd"}) # tds lmfao IYKYK
    for td in tds:
        for temp in td.span.get_attribute_list("id"):
            if "SubjectCode" in temp:
                if td.text.strip('\n'):
                    sub_codes.add(td.text.strip('\n'))
            elif "SubjectName" in temp:
                if td.text.strip('\n'):
                    course_now = td.text.strip('\n')
                    sub_names.add(course_now)
            elif "Padagoggy" in temp: ## ffs who does two mistakes in a single word in the frickin official form/site. get some help cheruset. PS: I am always open for any job ;)
                sub = course_now + " - " + td.text.strip('\n')
                subs.append(sub)
                sub_abbr = ""
                for word in sub.split(" "):
                    sub_abbr += word[0]
                subs_abbr.append(sub_abbr)
```

```python
get_subs_info(soup) # Trial

print('sub_codes: ', sub_codes, end='\n\n')
print('sub_names: ', sub_names, end='\n\n')
print('subs: ', subs, end='\n\n')
print('subs_abbr: ', subs_abbr, end='\n\n')
```

```python
def get_grades_info(soup):
    grades = []
    for grade in soup.find_all("table", {"id": "uclGrd1_grdResult"})[0].find_all("td", {"class": "GrdCel"})[1::2]:
        grades.append(grade.span.text)

    for i, sub in enumerate(subs_abbr):
        student_info[sub] = grades[i]
```

```python
get_grades_info(soup) # Trial
student_info
```

```python
def get_sgpa(soup):
    student_info['SGPA'] = soup.find("span", {"id": "uclGrd1_lblSGPA"}).text
```

```python
get_sgpa(soup) # Trial
student_info
```

```python
# Main function for getting everything together with loop. This gunna take a good amount of time because getting response from the request takes a good bit of time and then theres also the processing into the soup
for id_num in range(1, total_number_of_students):
    if id_num in gone_students_numbers:
        continue
    try:
        soup = get_html_response_soup(id_num)
        get_basic_student_info(soup)
        get_subs_info(soup)
        get_grades_info(soup)
        get_sgpa(soup)
        df = pd.concat([df, pd.DataFrame.from_records([student_info])])
    except:
        print("Had some issue with id: ", id_num)
```
